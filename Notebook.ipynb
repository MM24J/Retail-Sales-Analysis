```python
sales['Date'] = pd.to_datetime(sales['Date'],format='mixed')
sales = sales.sort_values(["Date"])
sales
```

```python
sales.info()
```

```python
features["CPI"] = features["CPI"].round(3)
features = features.loc[(features["Date"] <= "2012-12-10")]
features_df = features.rename(columns = {"Temperature":"Temp_Fahrenheit", "Fuel_Price":"Fuel_Price_$",
                                        "CPI":"Consumer_Price_Index_CPI", "Unemployment": "Unemployment_Pct"})
columns_to_drop = ["MarkDown1", "MarkDown2", "MarkDown3", "MarkDown4", "MarkDown5"]
features_df = features_df.drop(columns_to_drop, axis = 1)

features_df = features_df.sort_values(["Date"])
features_df
```

```python
sales_df = sales.rename(columns = {"Weekly_Sales":"Weekly_Sales_$"})
sales_df = sales_df.sort_values(["Date"])
sales_df.tail(10)
```

```python
stores_df = stores.rename(columns = {"Size":"Size_Sq_ft"})
stores_df
```

```python
retail_stores = features_df.merge(stores_df, on='Store', how='left')
columns_to_drop = ["IsHoliday", "Type"]
retail_stores = retail_stores.drop(columns_to_drop, axis=1)
retail_stores
```

```python
retail_stores['Date'] = retail_stores['Date'].dt.strftime('%m/%d/%Y')
retail_stores.head()
```

```python
retail_stores_df1 = retail_stores
```

```python
retail_stores_df1.to_csv("Resources/retail_stores_df1.csv", index=False)
```
